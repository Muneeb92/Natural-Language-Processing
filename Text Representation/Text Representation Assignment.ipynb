Practice converting a dataset into structured format, using the dataset from UCI repository about hotel reviews (https://archive.ics.uci.edu/ml/datasets/Eco-hotel) . Structure it with Tfidf scheme, removing noise words having doc frequency below 20 and stopwords having doc frequency above 200. 

```python
corpus = open('D:\\hotel.csv').read() 
```

```python
docs = corpus.split()
```

```python
docs
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vec = TfidfVectorizer(min_df = 20, max_df = 200) #Making obj of TfidfVectorizer 
```

```python
inputVec = vec.fit_transform(docs)
```
